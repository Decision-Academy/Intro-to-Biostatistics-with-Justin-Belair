## Percentiles and Quantiles

In NumPy, `np.percentile()` computes percentiles (input as 0–100) and `np.quantile()` computes quantiles (input as 0–1).

```python
import numpy as np
```

```python
salary = np.array([1000, 2000, 3000, 4000, 5000])

# Compute several quantiles (0th, 1st, 2nd, 3rd, 25th, 50th, 95th percentiles)
probs = [0, 0.01, 0.02, 0.03, 0.25, 0.5, 0.95]
quantiles = np.quantile(salary, probs)

for p, q in zip(probs, quantiles):
    print(f"{p*100:5.1f}th percentile: {q:.2f}")
```

## Different Quantile Algorithms

R supports 9 quantile algorithms via the `type` argument. NumPy provides equivalent methods via the `method` parameter.

| R `type` | NumPy `method` |
|----------|----------------|
| 6 | `'weibull'` |
| 7 (default) | `'linear'` (default) |
| 8 | `'median_unbiased'` |

```python
# Equivalent to R's type = 6
quantiles_type6 = np.quantile(salary, probs, method='weibull')
print("Type 6 (Weibull):")
for p, q in zip(probs, quantiles_type6):
    print(f"  {p*100:5.1f}%: {q:.2f}")
```

```python
# Equivalent to R's type = 8
quantiles_type8 = np.quantile(salary, probs, method='median_unbiased')
print("Type 8 (median_unbiased):")
for p, q in zip(probs, quantiles_type8):
    print(f"  {p*100:5.1f}%: {q:.2f}")
```
