## Variance and Standard Deviation

The variance and standard deviation measure the **spread** (dispersion) of data around the mean.

We use `ddof=1` in NumPy to get the **unbiased** (sample) estimator with $n-1$ in the denominator, matching R's `var()` and `sd()` defaults.

```python
import numpy as np
```

```python
salary2 = np.array([1000, 2000, 3000, 4000, 5000])

# Sample variance (ddof=1 for unbiased estimator)
print("Variance:", np.var(salary2, ddof=1))

# Sample standard deviation
print("Std Dev: ", np.std(salary2, ddof=1))
```

```python
# Confirm that std is simply the square root of the variance
print("sqrt(var):", np.sqrt(np.var(salary2, ddof=1)))
```

## Note on `ddof` in NumPy

- NumPy's **default** (`ddof=0`) uses $n$ as the denominator (biased / population estimator).
- Setting `ddof=1` uses $n-1$ (Bessel's correction), giving the **unbiased sample** estimator.
- R's `var()` and `sd()` use $n-1$ by default.
- Always set `ddof=1` when computing sample statistics to match R.
